```python
import pandas as pd
import jieba
from itertools import combinations
from collections import Counter

df = pd.read_excel("Merge文件.xlsx")
texts = df["微博内容"].dropna().astype(str).tolist()

stopwords = {"的","了","在","是","就","都","和","与","也","对","中","为","有","我","我们","10","11","12","20","收起","视频","微博","其中"}

counter = Counter()
for text in texts:
    tokens = [w for w in jieba.lcut(text) if len(w) >= 2 and w not in stopwords]
    uniq = set(tokens)
    for w1, w2 in combinations(sorted(uniq), 2):
        counter[(w1, w2)] += 1

top80 = counter.most_common(150)
for (w1, w2), cnt in top80:
    print(f"{w1}\t{w2}\t{cnt}")
```
